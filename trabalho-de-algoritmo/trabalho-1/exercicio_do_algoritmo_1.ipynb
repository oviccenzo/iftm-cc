##exercicio-3

```python
# Entrada de dados
numero_a = int(input("Digite o número A: "))
numero_b = int(input("Digite o número B: "))

# Inicialização das variáveis
contador = numero_a
resultado = numero_a

# Loop para somar o intervalo
while (contador <= numero_b):
    if (contador == numero_a):
        # Se for o primeiro número, apenas pula para o próximo
        contador = contador + 1
    else:
        # Soma o valor atual do contador ao resultado total
        resultado = resultado + contador
        contador = contador + 1

# Exibe o valor final da soma
print("Resultado: ", resultado)
```
